```python
def convert_to_gbp(num):
    return num * 0.79

nike['Close'].apply(convert_to_gbp)
```

```python
nike
```

```python
nike['GBP Close'] = nike['Close'].apply(convert_to_gbp)
```

```python
nike
```

```python
def volume_symbol(volume):
    if volume < 4000000:
        return 'Low'
    elif volume < 6000000:
        return 'Mid'
    else:
        return 'High'
```

```python
nike['Volume Symbol'] = nike['Volume'].apply(volume_symbol)
```

```python
nike
```

```python
def double_num(num):
    return num * 2
```

```python
nike['GBP Close'].apply(lambda gbp: gbp*2)
```

```python
def calculate_high_low_dif(high, low):
    return high - low
```

```python
nike['High Low Dif'] = nike[['High', 'Low']].apply(lambda cols: calculate_high_low_dif(cols['High'], cols['Low']), axis=1)
```

```python
nike
```

```python
nike['High Low Vectorize'] = np.vectorize(calculate_high_low_dif)(nike['High'], nike['Low'])
```
