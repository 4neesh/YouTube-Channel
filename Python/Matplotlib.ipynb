```python
a = np.linspace(0,10,11)
b = a ** 2

x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
axes = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.45,0.4,0.4])
axes.plot(a,b)
axes_2.plot(x,y)
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
axes = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.45,0.4,0.4])
axes.plot(a,b)
axes_2.plot(x,y)

axes.set_title('My dual axis')
axes.set_xlim(0,10)
axes.set_ylim(0,100)
axes.set_xlabel('Axes x label')
axes.set_ylabel('Axes y label')

axes_2.set_xlim(0,10)
axes_2.set_ylim(0,20)
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
axes = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.45,0.4,0.4])
axes.plot(a,b)
axes_2.plot(a,b)

axes.set_title('My dual axis')
axes.set_xlim(0,10)
axes.set_ylim(0,100)
axes.set_xlabel('Axes x label')
axes.set_ylabel('Axes y label')

axes_2.set_xlim(0,2)
axes_2.set_ylim(0,10)
```

```python
x = np.linspace(0,11,10)

fig = plt.figure(figsize=(3,2), dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x vs x^2')
ax.plot(x,x, label='x vs x')
ax.plot(x,x**3, label='x vs x^3')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlim(0,12)
plt.ylim(0,150)
ax.plot(x, x**2)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlim(0,12)
plt.ylim(0,150)
ax.plot(x, x**2, color='purple', label='x vs x^2')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlim(0,12)
plt.ylim(0,150)
ax.plot(x, x**2, color='purple', label='x vs x^2', lw='2', ls='-', marker='x', ms='10', markeredgecolor='blue')
ax.plot(x,x, color='orange', label='x vs x', lw='1', ls='-', marker='o', markeredgecolor='black', markerfacecolor='green')
ax.legend()
```
